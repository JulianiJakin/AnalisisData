## Visualization & Explanatory Analysis

### Pertanyaan 1:
Bagaimana Performa Pemesanan Bike Sharing dalam Beberapa Bulan Terakhir (2012)?

```python
all_df['dteday_x'] = pd.to_datetime(all_df['dteday_x'])

all_df.set_index('dteday_x', inplace=True)

all_df_2012 = all_df[all_df.index.year == 2012]
monthly_orders_df = all_df_2012.resample(rule='M').agg({
    "instant": "nunique", 
})

monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%m')
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "dteday_x": "order_date",
    "instant": "order_count"
}, inplace=True)
monthly_orders_df.head()
```

```python
all_df_2012 = all_df[all_df.index.year == 2012]

monthly_orders_df = all_df_2012.resample(rule='M').agg({
    "instant": "nunique",
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%B' '%Y') #mengubah format order date menjadi nama bulan
 
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "dteday_x": "order_date",
    "instant": "order_count"
}, inplace=True)
 
plt.figure(figsize=(20, 5)) 
plt.plot(monthly_orders_df["order_date"], monthly_orders_df["order_count"], marker='o', linewidth=2, color="#72BCD4") 
plt.title("Number of Orders per Month (2012)", loc="center", fontsize=20) 
plt.xticks(fontsize=10) 
plt.yticks(fontsize=10) 
plt.show()
```

### Pertanyaan 2:
Pada Musim Apa Pelanggan Paling Sering Melakukan Pemesanan Bike Sharing?

```python
season_mapping = {1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"}
all_df['season'] = all_df['season_x'].map(season_mapping)

seasonal_orders = all_df.groupby('season').size().reset_index(name='order_count')

print(seasonal_orders)
```

```python
season_mapping = {1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"}
all_df['season'] = all_df['season_x'].map(season_mapping)

season_orders = all_df.groupby('season').size()

plt.figure(figsize=(10, 5))
season_orders.plot(kind='bar', color="#72BCD4")
plt.title('Number of Orders by Season', fontsize=16)
plt.xlabel('Season', fontsize=14)
plt.ylabel('Number of Orders', fontsize=14)
plt.xticks(rotation=0)
plt.show()
```
